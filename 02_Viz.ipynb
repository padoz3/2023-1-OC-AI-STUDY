````python
import matplotlib.pyplot as plt

X = ["Tom", "Amy", "Jim", "Bill", "Pegi"]
Y = [8, 23, 52, 37, 64]

fig, ax = plt.subplots(figsize = (8,6))

bars = ax.bar(X, Y, color = "lightgray", edgecolor = "gray", zorder = 17)

for a in bars:
    x = a.get_x() + a.get_width() / 2
    y = a.get_height()
    ax.annotate(str(y), xy = (x, y + 4), ha = "center", color = "black", zorder = 1)

ax.axhline(y = 45, linestyle = "--", color = "blue")
ax.axhline(y = 15, linestyle = "--", color = "red")

ax.axhspan(0, 15, facecolor = "lightcoral", alpha = 0.5, zorder = -1)

label_fontdict = {"fontsize": 14, "fontweight": "bold", "color": "gray"}
ax.set_xlabel("Salesperson", fontdict = label_fontdict, labelpad = 12)
ax.set_ylabel("Revenue (1000 $)", fontdict = label_fontdict, labelpad = 12)

plt.tight_layout()

plt.show()
````
